# SALES **OVERVIEW**

```python
sales_asp = asp[asp['amount'] != 'N/A']
sales_asp.head()
```

```python
total_revenue = sales_asp['amount'].sum()
total_revenue
```

```python
total_orders = sales_asp['order_id'].nunique()
total_orders
```

```python
avg_ord_value = total_revenue / total_orders
avg_ord_value
```

```python
total_quantity = sales_asp['qty'].sum()
total_quantity

min_max_quantity = sales_asp['qty'].agg(['min', 'max'])
min_max_quantity
```

```python
## KPI's ##

print(f"Total Revenue:₹{total_revenue:,.2f}")
print(f"Total Orders: {total_orders:,}")
print(f"Average Order Value: ₹{avg_ord_value:,.2f}")
print(f"Total Quantity Sold: {total_quantity:,}",
      f"Minimum Quantity: {min_max_quantity['min']}",f"Maximum Quantity: {min_max_quantity['max']}")
```

```python
sales_asp['week'] = sales_asp['date'].dt.to_period('W')

weekly_sales = sales_asp.groupby('week')['amount'].sum().reset_index()

weekly_sales['week'] = weekly_sales['week'].dt.to_timestamp()

print("Weekly Sales:")
print(weekly_sales)
```

```python
sales_asp['month'] = sales_asp['date'].dt.to_period('M')


monthly_sales = sales_asp.groupby('month')['amount'].sum().reset_index()
monthly_sales

monthly_sales['month'] = monthly_sales['month'].dt.to_timestamp()


```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_sales['month'], monthly_sales['amount'])
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(weekly_sales['week'], weekly_sales['amount'])
plt.title('Weekly Sales Trend')
plt.xlabel('Weeks')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

# **FULFILLMENT ANALYSIS**

```python
fulfillment_asp = asp.copy()
```

```python
orders_by_fulfillment = (fulfillment_asp.groupby('fulfilled_by')['order_id']
                         .nunique()
                         .sort_values(ascending=False))
orders_by_fulfillment
```

```python
asp['status'].value_counts(dropna=False)
```

```python
cancelled_orders = fulfillment_asp[
    fulfillment_asp['status'].str.contains('cancelled', case=False, na=False)]

cancel_rate = (
    cancelled_orders
    .groupby('fulfilment')['order_id']
    .nunique()
    / fulfillment_asp.groupby('fulfilment')['order_id'].nunique()
) * 100

cancel_rate
```

```python
len(cancelled_orders)
```

```python
cancel_rate
```

```python
plt.figure(figsize=(10,5))
plt.pie(cancel_rate, labels=cancel_rate.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Cancellation : Amazon vs Merchant')
plt.show()

```

```python
plt.figure(figsize=(6,4))
plt.bar(['Amazon', 'Merchant'], cancel_rate)
plt.title('Cancellation Rate by Fulfilment Method')
plt.xlabel('Fulfilment Method')
plt.ylabel('Cancellation Rate (%)')
plt.tight_layout()
plt.show()
```

# **GEOGRAPHICAL ANALYSIS**

```python
geo_asp = sales_asp.copy()
```

```python
state_revenue = (geo_asp
                 .groupby('ship_state')['amount']
                 .sum()
                 .sort_values(ascending=False))
state_revenue.head(10)
```

```python
state_revenue_share = (state_revenue/total_revenue)*100
state_revenue_share.head(10)
top_10_states  = state_revenue.head(10)
```

```python
plt.figure(figsize=(10,5))
top_10_states.plot(kind='bar')
plt.title('Top 10 States by Revenue')
plt.xlabel('State')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
city_revenue = (geo_asp
                .groupby('ship_city_cleaned')['amount']
                .sum()
                .sort_values(ascending=False))
city_revenue
city_revenue_no_other = city_revenue.drop('Other')
top_10_cities = city_revenue_no_other.head(10)
top_10_cities
```

```python
cit_revenue_share = (top_10_cities/total_revenue)*100
cit_revenue_share
```

```python
plt.figure(figsize=(10,5))
top_10_cities.plot(kind='bar')
plt.title('Top 10 Cities by Revenue')
plt.xlabel('Top 10 Cities')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
